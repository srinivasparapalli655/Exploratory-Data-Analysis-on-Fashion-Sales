```python
def detect_outliers_iqr(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    #outliers_df = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return lower_bound, upper_bound
```

```python
low,high= detect_outliers_iqr(df,'units_sold')
print('lower bound: ',low)
print('upper bound: ',high)
```

```python
from scipy.stats import shapiro, norm
import seaborn as sns
import matplotlib.pyplot as plt

def check_normal_distribution(column):
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True, bins=30, stat='density')
    mean = df[column].mean()
    std = df[column].std()
    x = np.linspace(df[column].min(), df[column].max(), 100)
    plt.plot(x, norm.pdf(x, mean, std), 'r', label='Normal PDF')
    plt.title(f"{column} Distribution vs Normal Curve")
    plt.legend()
    plt.show()

    stat, p = shapiro(df[column])
    print(f"{column} - Shapiro-Wilk Test p-value: {p:.4f} {'(Normal)' if p > 0.05 else '(Not Normal)'}")

# Example:
check_normal_distribution('profit')
```

- profit looks close to normal.
- units_sold, unit_price, and sales_amount are positively skewed and fail normality test.

```python
# Apply log transformation and plot
df['log_sales'] = np.log1p(df['sales_amount'])

plt.figure(figsize=(8, 4))
sns.histplot(df['log_sales'], kde=True, bins=30)
plt.title("Log-Transformed Sales Amount Distribution")
plt.show()
```

- After log transformation, sales_amount appears more symmetric.
